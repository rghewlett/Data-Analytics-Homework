```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
city_data

ride_data = pd.read_csv(ride_data_to_load)
ride_data

# Combine the data into a single dataset
merge_table = pd.merge(city_data, ride_data, on = "city")
merge_table

# Display the data table for preview
merge_table.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
# Urban Cities Plot dataframe
urban_df = merge_table.loc[(merge_table.type == "Urban")]
urban_df

# Urban Cities x-axis: unique rides in each urban city
urban_rides = urban_df.groupby("city").ride_id.count()
urban_rides

# Urban Cities y-axis
avg_fare_urb = urban_df.groupby("city").fare.mean()
avg_fare_urb

# Plot point sizing based on how many drivers
driver_ct_urban = urban_df.groupby("city").driver_count.mean()

# Plot Urban Cities Data Analysis
urban = plt.scatter(urban_rides, avg_fare_urb, facecolors="coral", alpha=0.75, edgecolor="black", linewidth=1.0, s=driver_ct_urban*10, label="Urban")

#------------------------------------------------------------------------------------------------------------------

# Rural Cities Plot dataframe
rural_df = merge_table.loc[(merge_table.type == "Rural")]
rural_df

# Rural Cities x-axis: unique rides in each rural city
rural_rides = rural_df.groupby("city").ride_id.count()
rural_rides

# Rural Cities y-axis: average fare for rides in each rural city
avg_fare_rural = rural_df.groupby("city").fare.mean()
avg_fare_rural

# Plot point sizing based on how many drivers
driver_ct_rural = rural_df.groupby("city").driver_count.mean()
driver_ct_rural

# Plot Rural Cities Data Analysis
rural = plt.scatter(rural_rides, avg_fare_rural, facecolors="Gold", alpha=0.75, edgecolor="black", linewidth=1.0, s=driver_ct_rural*10, label="Rural")

#--------------------------------------------------------------------------------------------------------------------

# Suburban Cities Plot dataframe
suburban_df = merge_table.loc[(merge_table.type == "Suburban")]
suburban_df

# Suburban Cities x-axis: unique rides in each suburban city
suburban_rides = suburban_df.groupby("city").ride_id.count()
suburban_rides

# Suburban Cities y-axis: average fare for rides in each suburban city
avg_fare_sub = suburban_df.groupby("city").fare.mean()
avg_fare_sub

# Plot point sizing based on how many drivers 
driver_ct_sub = suburban_df.groupby("city").driver_count.mean()
driver_ct_sub

# Plot Suburban Cities Data Analysis
suburban = plt.scatter(suburban_rides, avg_fare_sub, facecolors="lightskyblue", alpha=0.75, edgecolor="black", linewidth=1.0, s=driver_ct_sub*10, label="Suburban")

# Formatting
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.text(-1,1,"Note",horizontalalignment='center',verticalalignment='center')
plt.grid(True)
plt.xlim()
plt.legend(title = "City Types", handles=[urban, suburban, rural])

# Text Label
plt.text(42,35,"Note: \n Circle size correlates with driver count per city.", fontsize=12)

# Save Figure
plt.savefig("Bubble_Plot_.png", bbox_inches = "tight")

# Show Plot
plt.show()

#------------------------------------------------------------------------------------------------------------------

```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

# Build Pie Chart

# Save Figure

#-----------------------------------------------------------------------------------------------------------------

# Calculate total fares
total_fares = merge_table.groupby("type").fare.sum()
total_fares

# Create pie chart labels
labels = ["Rural", "Suburban", "Urban"]

# Create values of pie chart slices
sizes = [4327.93, 19356.33, 39854.38]

# The colors of each section of the pie chart
colors = ["Gold", "lightskyblue", "lightcoral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)

# Automatically finds the percentages of each part of the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

# Formatting
plt.title("% of Total Fares by City Type")
plt.axis("tight")

# Show Figure
plt.show()

# Save Figure
plt.savefig("Percentage_of_Total_Fares_by_City_Type.png")
```
